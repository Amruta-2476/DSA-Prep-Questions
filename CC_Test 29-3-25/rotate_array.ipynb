```python
N, K = map(int, input().split())
# print(N, K)
arr = list(map(int, input().split()))
# print(arr)

K = K%N

for _ in range(K):
  last = arr[-1]
  for i in range(N-1, 0, -1):
    arr[i] = arr[i-1]
  arr[0] = last

for i in range(N):
  if i==N-1:
    print(arr[i])
  else:
    print(arr[i], end=' ')
```
