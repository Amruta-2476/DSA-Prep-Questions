using expand around center approach with O(n) time complexity, which is optimal for this problem.

Solution Explanation <br>
Expand Around Center: Each palindrome has a center. It can be either a single character (odd-length palindrome) or between two characters (even-length palindrome).<br>
We expand outward from each center and check for the longest palindrome.

Steps:<br>
Loop through each character in s.<br>
Try to expand around odd-length and even-length centers.<br>
Keep track of the longest palindrome found.<br>

```python
class Solution:
    def longestPalindrome(self, s):
        """
        :type s: str
        :rtype: str
        """
        if not s or len(s) == 1:
            return s  # If empty or single character, return as is
        
        def expandAroundCenter(left, right):
            """Expand from center and return longest palindromic substring"""
            while left >= 0 and right < len(s) and s[left] == s[right]:
                left -= 1
                right += 1
            return s[left + 1:right]  # Return valid palindrome substring
        
        longest = ""
        
        for i in range(len(s)):
            # Odd-length palindrome centered at i
            odd_palindrome = expandAroundCenter(i, i)
            if len(odd_palindrome) > len(longest):
                longest = odd_palindrome
            
            # Even-length palindrome centered between i and i+1
            even_palindrome = expandAroundCenter(i, i + 1)
            if len(even_palindrome) > len(longest):
                longest = even_palindrome
        
        return longest


# Taking input from the user
s = input().strip()

# Creating an instance of Solution and calling the method
solution = Solution()
result = solution.longestPalindrome(s)

# Printing the result
print(result)
```
