```python
def largestRectangle(heights):
    n = len(heights)
    
    # Arrays to store left and right boundaries
    left = [0] * n
    right = [0] * n
    
    # Find left boundary for each building
    for i in range(n):
        p = i - 1
        while p >= 0 and heights[p] >= heights[i]:
            p = left[p]  # Jump to the left boundary
        left[i] = p  # Store left boundary
    
    # Find right boundary for each building
    for i in range(n - 1, -1, -1):
        p = i + 1
        while p < n and heights[p] >= heights[i]:
            p = right[p]  # Jump to the right boundary
        right[i] = p  # Store right boundary
    
    # Calculate max area using the boundaries
    max_area = 0
    for i in range(n):
        width = right[i] - left[i] - 1  # Width of the rectangle
        max_area = max(max_area, heights[i] * width)
    
    return max_area

# Example usage
heights = [1, 2, 3, 4, 5]
print(largestRectangle(heights))  # Output: 9
```
