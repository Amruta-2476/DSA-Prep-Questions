```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
# queue using 2 stacks : 
# NOTE = stack1 is for enqueue and stack2 is for dequeue. copy s1 reverse into s2


class QueueUsingStacks():
    def __init__(self):
        self.s1 = []
        self.s2 = []
    
    def enqueue(self, val):
        self.s1.append(val)
    
    def dequeue(self):
        if len(self.s2) == 0: 
            while self.s1:
                s1_element = self.s1.pop()
                self.s2.append(s1_element)
        self.s2.pop()
        
    def printFront(self):
        if len(self.s2) == 0: 
            while self.s1:
                s1_element = self.s1.pop()
                self.s2.append(s1_element)
        print(self.s2[-1])
        
        


q = int(input())  # no.of queries
queue = QueueUsingStacks()

for i in range(q):
    query = list(map(int, input().split()) )   
    if query[0] == 1:
        queue.enqueue(query[1])  
    elif query[0] == 2:
        queue.dequeue() 
    elif query[0] == 3:
        queue.printFront() 
```
